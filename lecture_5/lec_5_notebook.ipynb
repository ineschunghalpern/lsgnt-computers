```python
def factorial (n):
    if n==0:
        return 1
    return n*factorial(n-1)

factorial(3)
```

```python
fib = {0:1, 1:1}

def fibonnacci(n):
    if n==0:
          return 1
    if n==1:
     return 1
    elif n in fib:
        return fib[n]
    else:
        fib[n] = fibonnacci(n-1) + fibonnacci(n-2)
        return fib[n ]
```

```python
def fyshuffle(list):
    n = len(list)
    for i in range(n-1):
        j = random.randint(i, n-1)
        holder = list[i]
        list[i] = list[j]
        list[j] = holder
    return list
```
